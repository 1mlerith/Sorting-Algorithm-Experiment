```python
import matplotlib.pyplot as plt

# --- Data hasil eksperimen ---
data_size = [10, 50, 100, 500, 1000, 2000, 4000, 8000, 10000, 16000]
bubble_time = [0, 0, 1, 15, 12, 15, 71, 286, 485, 1410]
insertion_time = [0, 0, 0, 0, 2, 4, 17, 83, 137, 342]

# --- Plot garis ---
plt.figure(figsize=(8, 5))
plt.plot(data_size, bubble_time, marker='o', label='Bubble Sort')
plt.plot(data_size, insertion_time, marker='s', label='Insertion Sort')

# --- Tambahan label dan judul ---
plt.title("Perbandingan Waktu Eksekusi Sorting (Linked List)")
plt.xlabel("Ukuran Data (n)")
plt.ylabel("Waktu Eksekusi (ms)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# --- Tampilkan ---
plt.show()
```
